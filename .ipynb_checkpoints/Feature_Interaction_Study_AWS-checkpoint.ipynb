# Features Interactions Study on the Kaggle Santander Challenge
### *Yifan Xie*

This Jupyter Notebook illustrates a feature selection technique used to identify useful feature interactions - potential features that can be combined to create useful new features. This technique is particularly useful when dataset provided for machine learning purpose comes with largely anonymous or encrypted features, or when the circumstance of the task does not allow application of domain knowledge. 

The dataset used in this illustration is that of the Kaggle Challenge [Santander Customer Satisfaction Challenge](https://www.kaggle.com/c/santander-customer-satisfaction), in which the author is currently ranked 639th out of 3852 teams at the time of writing (13/04/2016)

In the following text, the term **elemental feature** refer to original features that comes with the dataset, while the term **interactive feature** refer to features that are created to capture the interaction between elemental feature. 

The interactive features are created in the following manner: for a given list elemental features **[feature 1, feature 2,... feature n]**, a feature of the form **"feature 1-feature 2-...-feature n"** is created to capture the interaction of the aforementioned list of elemental features. For illustration purpose, feature interaction with up to four elemental features are studied (i.e. n<=4).  Also for this purpose, only subtraction operation is performed while in other mathmetical operation are aften considered in operation/competition scenario. 

Overall speaking, the script provided here shows the following functional execution :

1. **Feature Creation:** Create n-way interactive features from pre-selected pairs, triplets, and quadruplets of elemental features

2. **Interactive Feature Evaluation:** For each n-way interactive features created in step 1, perform the following:
     1. Use a simple learner to perform evaluation on each elemental feature of the concerned interactive feature
     2. Use the same simple learner to perform evaluation on the concerned interactive feature
     3. Compare the contribution of the interactive features in score function optimisation to those of the elemental features
     
3. **Overall Feature Evaluation:** Combine promising interactive features with pre-defined list of original feature that contribute significantly to cost function, and perform overall single feature ranking based on score function optimisation

4. **Random Feature Set Evaluation:** Perform random feature selection to create 100 random set of 50 features from 150 best single features (elemental or interactive).


The learner that is used here is [ExtremeBoosting](https://github.com/dmlc/xgboost), an implementation of gradient boost tree. The pre-defined pairs, triplets and quadruplets of elemental features, and the prededinfed list of important elemental features are identified using package [Xgbfi](https://github.com/Far0n/xgbfi). 

Moreover, the feature selection part of this note book is partly inspired and based on the [Advanced Feature Exploration](https://www.kaggle.com/selfishgene/santander-customer-satisfaction/advanced-feature-exploration/discussion) shared on Kaggle's [competition shared scripts](https://www.kaggle.com/c/santander-customer-satisfaction/scripts)

The scoring function used in this competition is [Area under the ROC curve](https://www.kaggle.com/c/santander-customer-satisfaction/details/evaluation). For simplicity, the [auc](http://scikit-learn.org/stable/modules/generated/sklearn.metrics.auc.html) function provided by the library [sklearn](http://scikit-learn.org/stable/) is used.
 

## Preparation: Load dependency, data set, and simple data cleaning

```python
# Load dependency
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import cluster
from sklearn import cross_validation
from sklearn.metrics import roc_auc_score as auc
import time
import matplotlib.pyplot as plt
import xgboost as xgb
import pyexcel as pe
import pyexcel.ext.xlsx

#set plot size
plt.rcParams['figure.figsize'] = (10, 10)

#Load competition training data set 
trainDataFrame = pd.read_csv('./data/train.csv')

# remove constant columns
colsToRemove = []
for col in trainDataFrame.columns:
    if trainDataFrame[col].std() == 0:
        colsToRemove.append(col)
trainDataFrame.drop(colsToRemove, axis=1, inplace=True)

# remove duplicate columns
colsToRemove = []
columns = trainDataFrame.columns
for i in range(len(columns)-1):
    v = trainDataFrame[columns[i]].values
    for j in range(i+1,len(columns)):
        if np.array_equal(v,trainDataFrame[columns[j]].values):
            colsToRemove.append(columns[j])

            
# simple pre-processing to separate label from the training features
trainDataFrame.drop(colsToRemove, axis=1, inplace=True)
trainLabels = trainDataFrame['TARGET']
trainFeatures = trainDataFrame.drop(['ID','TARGET'], axis=1)

trainFeatures['30Div15']=trainFeatures['saldo_var30']/trainFeatures['var15']
# keep a record of original features of the train data, this is a convinient reference for usage further down the script
org_cols=trainFeatures.columns
```

```python
trainLabels.shape
```

## Step 1: Feature Creation
In this step, the predefined pairs, triplets and quadruplets features are loaded.
For each valid pair/triplet/quadruplets that don't have duplicated elemental features, an interactive feature is created using subtraction operation. 

```python
#load predifined candidate list of elemental features 
book = pe.get_book(file_name="./data/XgbFeatureInteractions_zero_dup_removed.xlsx")
sheets = book.to_dict()

#function to create interactive feature. 
def create_interact_features(feat_n_way, features):
    # feat_n_way: list of elemental features
    # features: input training features 
    
    #initialise list variable for return value 
    interact_feat=[]
    nway_list=[]
    for i in range(len(feat_n_way)):
        feat_str = feat_n_way[i]
        feat_list=feat_str.split('|')  
        
        # if the number of unique value in the list is equal the number of unique value
        # then it means there is no duplicated value in the list.
        # in this case the list of elemental features being looked at is considered valid for this operation
        if len(np.unique(feat_list))==len(feat_list):
            temp_featname=""

            #recursively create the interactive feature name, 
            #and construct the interactive feature value via subtraction
            for feat_id in range(len(feat_list)):
                if len(temp_featname)==0:
                    temp_featname=feat_list[feat_id]
                    temp_feat=features[feat_list[feat_id]]
                else:
                    temp_featname=temp_featname + "-" + feat_list[feat_id]
   
                    temp_feat=temp_feat-features[feat_list[feat_id]]
  
            features[temp_featname]=temp_feat
            interact_feat.append(temp_featname)
            nway_list.append(feat_list)
            
    #return the list 
    return(interact_feat, features, nway_list)
            
#read list of pairs, triplets and quadruplets of elemental features from soure file
feat_pair_list = np.array(sheets['Interaction Depth 1'])[1:, 0]
feat_triplet_list = np.array(sheets['Interaction Depth 2'])[1:, 0]
feat_quadruplet_list = np.array(sheets['Interaction Depth 3'])[1:, 0]

# call the 'create_interact_features' function to create interactive features
feat_two_way, trainFeatures, feat_pair=create_interact_features(feat_pair_list, trainFeatures)
feat_three_way, trainFeatures, feat_triplet=create_interact_features(feat_triplet_list, trainFeatures)
feat_four_way, trainFeatures, feat_quadruplet=create_interact_features(feat_quadruplet_list, trainFeatures)
```

## Step 2: Feature Evaluations
The *"interactive_feature_comparision"* function use a simple gradient boosting tree learner to perform evaluation for each interactive feature, and its associate elemental features

```python
#function to evaluate the performance of interactive feature and each associated elemental features 
def interactive_feature_comparision(feat_n_way, elemental_features, features, labels):
    startTime = time.time()

    #list for feature and auc columns creation
    feat_cols=[]
    AUC_cols=[]

    #perform a train/validation split on the training data
    X_train, X_valid, y_train, y_valid = cross_validation.train_test_split(features, labels,
                                                                       test_size=0.5, random_state=1)
    
    #define a simple learner to perform the evaluation. 
    verySimpleLearner = xgb.XGBClassifier(max_depth=3, learning_rate=0.1, n_estimators=20,
                                            subsample=0.6, seed=1234)
    
    
    #define all the columns names required by the feature comparison table
    for i in range(len(elemental_features[0])):
        temp_feat_col="feature"+str(i+1)
        temp_AUC_col="AUC"+str(i+1)
        feat_cols.append(temp_feat_col)
        AUC_cols.append(temp_AUC_col)
    feat_cols.append("interactive_feature")    
    feature_comparison_table = pd.DataFrame(index=range(len(feat_n_way)), 
                                            columns=feat_cols + AUC_cols+['AUC_interact', 'Diff'])
    
    
    # loop through each item in the input list of n way features
    #calculate the AUC value for each elemental feature and the interactive fature
    
    for k, feature in enumerate(feat_n_way):
        #print progressive message
        if (k+1)/20==int((k+1)/20):
            print("%d features have been evaluated" %(k+1))
            
        for i in range(len(feat_cols)):
            feat_col_name=feat_cols[i]
            if i<(len(feat_cols)-1):
                auc_col_name='AUC'+str(i+1)
                feat_name=elemental_features[k][i]
            else:
                feat_name=feat_n_way[k]
                auc_col_name='AUC_interact'

            trainInputFeature = X_train[feat_name].values.reshape(-1,1)
            validInputFeature = X_valid[feat_name].values.reshape(-1,1)
            verySimpleLearner.fit(trainInputFeature, y_train)

            validAUC = auc(y_valid, verySimpleLearner.predict_proba(validInputFeature)[:,1])
            feature_comparison_table.ix[k,feat_col_name] = feat_name
            feature_comparison_table.ix[k,auc_col_name] = validAUC  

    feature_comparison_table['Diff']=feature_comparison_table['AUC_interact']- \
                                        feature_comparison_table[AUC_cols].max(axis=1)

    
    feature_comparison_table_sorted=feature_comparison_table.sort_values(by='Diff', 
                                                                     axis=0, 
                                                                     ascending=False).reset_index(drop=True)
    

    print("finished evaluating interactive features. took %.2f minutes" %((time.time()-startTime)/60))
    return feature_comparison_table_sorted
```

### Step 2.1 Perform evaluation on two-way interactive features
Firstly, we call the *"interactive_feature_comparison"* function to generate the comparison table for all the two-way interactive features.

```python
two_way_features_comparison=interactive_feature_comparision(feat_two_way, feat_pair, trainFeatures, trainLabels)
```

##Step 3: Overall Feature Evaluation

First of all, we need to put all the "strong" interactive features(i.e. features that improved on their elemental features by more than 0.02) together, and then combine them with the original training features.

```python
#extract from the features comparison dataframe all the strong interactive features
filter_value=0.04

strong_two_way_features=two_way_features_comparison[two_way_features_comparison['Diff']>filter_value]['interactive_feature']
strong_three_way_features=three_way_features_comparison[three_way_features_comparison['Diff']>filter_value]['interactive_feature']
strong_four_way_features=four_way_features_comparison[four_way_features_comparison['Diff']>filter_value]['interactive_feature']


#Consolidate all the strong interactive features
strong_features=strong_two_way_features.append(strong_three_way_features).append(strong_four_way_features)

#Consolidate the original features with the strong interactive features. 
# Overall_trainFeatures=pd.concat([trainFeatures[org_cols], trainFeatures[strong_features]], axis=1, 
#                                  join_axes=[trainFeatures.index])

Overall_trainFeatures=trainFeatures[org_cols]
```

Next step is to create a function *individual_feature_evaluation* that can perform individual feature evaluation on the overall feature set. This function is very similar to the *interactive_feature_comparision* defined in step 2, but this time we no longer distinguish interactive features and elemental features.

```python
def individual_feature_evaluation(features, labels):
    startTime = time.time()
    feat_cols=[]
    AUC_cols=[]

    X_train, X_valid, y_train, y_valid = cross_validation.train_test_split(features, labels,
                                                                       test_size=0.5, random_state=1)

    verySimpleLearner = xgb.XGBClassifier(max_depth=3, learning_rate=0.1, n_estimators=20,
                                            subsample=0.6, seed=1234)

    startTime = time.time()
    singleFeatureTable = pd.DataFrame(index=range(len(X_train.columns)), columns=['feature','AUC'])
    for k,feature in enumerate(X_train.columns):
        if (k+1)/50==int((k+1)/50):
            print("%d features have been evaluated" %(k+1))
            
        trainInputFeature = X_train[feature].values.reshape(-1,1)
        validInputFeature = X_valid[feature].values.reshape(-1,1)
        verySimpleLearner.fit(trainInputFeature, y_train)

        validAUC = auc(y_valid, verySimpleLearner.predict_proba(validInputFeature)[:,1])
        singleFeatureTable.ix[k,'feature'] = feature
        singleFeatureTable.ix[k,'AUC'] = validAUC
    
    print("finished evaluating single features. took %.2f minutes" %((time.time()-startTime)/60))

    #Show single feature AUC performace
    singleFeatureTable = singleFeatureTable.sort_values(by='AUC', axis=0, ascending=False).reset_index(drop=True)
    
    return singleFeatureTable
    
```

Now, let's call the *individual_feature_evaluation* function to perform evaluation on each feature, and then we can have a look at the top 100 features by AUC score. As shown below, it is very interesting to see that most of the top features are interactive features which we created in steps 1 and 2.

```python
overall_feature_comparison=individual_feature_evaluation(Overall_trainFeatures, trainLabels)
print(overall_feature_comparison.ix[:50,:])
```

```python
#define the number of feature to be include in each feature set
numFeaturesInCombination = 250

#define the number of feature set to be randomly created
numCombinations = 500

#define the selection criteria of selecting best scored features
numBestSingleFeaturesToSelectFrom = 300


featuresToUse = overall_feature_comparison.ix[0:numBestSingleFeaturesToSelectFrom-1,'feature']
featureColumnNames = ['feature'+str(x+1) for x in range(numFeaturesInCombination)]
featureCombinationsTable = pd.DataFrame(index=range(numCombinations), columns=featureColumnNames + ['combinedAUC'])


#perform a train/validation split on the training data
# Overall_trainFeatures=Overall_trainFeatures.replace(0,np.nan).to_sparse()

X_train, X_valid, y_train, y_valid = cross_validation.train_test_split(Overall_trainFeatures, trainLabels, 
                                                                       test_size=0.5, random_state=1)
#define a simple learner to perform the evaluation. 
# simpleLearner = xgb.XGBClassifier(max_depth=5, learning_rate=0.1, n_estimators=120, colsample_bytree=0.4,
#                                 subsample=1, seed=1234)

simpleLearner = xgb.XGBClassifier(max_depth=5, learning_rate=0.1, n_estimators=50, colsample_bytree=0.4,
                                subsample=1, seed=1234)



startTime = time.time()
for combination in range(numCombinations):

    #print progressive message
    if (combination+1)/20==int((combination+1)/20):
        print("%d features combinations have been evaluated" %(combination+1))
    
    # generate random feature combination
    randomSelectionOfFeatures = sorted(np.random.choice(len(featuresToUse), numFeaturesInCombination, replace=False))

    # store the feature names
    combinationFeatureNames = [featuresToUse[x] for x in randomSelectionOfFeatures]
    for i in range(len(randomSelectionOfFeatures)):
        featureCombinationsTable.ix[combination,featureColumnNames[i]] = combinationFeatureNames[i]

    # build features matrix to get the combination AUC
    trainInputFeatures = X_train.ix[:,combinationFeatureNames]
    validInputFeatures = X_valid.ix[:,combinationFeatureNames]
    
    # train learner
    simpleLearner.fit(trainInputFeatures, y_train)
    
    # store AUC results
    validAUC = auc(y_valid, simpleLearner.predict_proba(validInputFeatures)[:,1])        
    featureCombinationsTable.ix[combination,'combinedAUC'] = validAUC
    
    
validAUC = np.array(featureCombinationsTable.ix[:,'combinedAUC'])
print("(min,max) AUC = (%.4f,%.4f). took %.1f minutes" % (validAUC.min(),validAUC.max(), (time.time()-startTime)/60))
```

```python
# show the histogram of the feature combinations performance 
plt.figure(); plt.hist(validAUC, 100, facecolor='blue', alpha=0.75)
plt.xlabel('AUC'); plt.ylabel('frequency'); plt.title('feature combination AUC histogram'); plt.show()

featureCombinationsTable = featureCombinationsTable.sort_values(by='combinedAUC', axis=0, 
                                                                ascending=False).reset_index(drop=True)
featureCombinationsTable.ix[:20,:]
```

## Conclusion

An exercise to automatically generate evaluate interactive features is demonstrated in this Jupyter Notebook, this type of techniques is particularly suitable for feature engineering when features are anonymous in the given dataset. 

Additionally, the random based feature selection approach allow for a relatively low-cost approach for feature selection which take both original features and interactive features into account.

Further improvement of this script would be the following:
1. Using clustering techniques to identify feature sets that provide good scoring and yet different from each other. For technical details, readers can refer to this [kaggle script](https://www.kaggle.com/selfishgene/santander-customer-satisfaction/advanced-feature-exploration/discussion).

2. The basic learner of in Step 3 can be replaced with a more well-tuned learner, and cross-validation techniques can be used to evaluate the score based on the whole data set. This would significantly increase the computation time, but shall yield more realistic results. 

3. Different ways to filter interactive features shall be explored, to achieve a more optimum solutions. 

4. It is also worth exploring different way to mix the original and interactive feature. For instance, one might want to preserve more original features even if their individual score is lower.
